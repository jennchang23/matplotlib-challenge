```python
# Dependencies and Setup
%matplotlib inline
import csv
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
from scipy import stats

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
drug_df = pd.read_csv(mouse_drug_data_to_load)
trial_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_df = pd.merge(trial_df, drug_df, on = "Mouse ID", how = "left")

# Display the data table for preview
combined_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
# Convert to DataFrame
tumorresponse_avg_df = combined_df.loc[:,["Drug","Timepoint","Tumor Volume (mm3)"]].groupby(["Drug","Timepoint"]).mean().reset_index().rename(columns = {"Tumor Volume (mm3)": "Tumor Volume (mm3) Average"})

# Preview DataFrame
tumorresponse_avg_df
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Convert to DataFrame
tumorresponse_stderr_df = combined_df.loc[:,["Drug","Timepoint","Tumor Volume (mm3)"]].groupby(["Drug","Timepoint"]).sem().reset_index().rename(columns = {"Tumor Volume (mm3)": "Tumor Volume (mm3) Std Error"})

# Preview DataFrame
tumorresponse_stderr_df.head()
```

```python
# Minor Data Munging to create a cross-tabulated dataframe -- for average tumor response to drugs

drugavgtumor_bytimepoint=pd.crosstab(index = tumorresponse_avg_df["Timepoint"], columns = tumorresponse_avg_df["Drug"], values = tumorresponse_avg_df["Tumor Volume (mm3) Average"], aggfunc='mean')

# Preview that Reformatting worked
drugavgtumor_bytimepoint
```

```python
# crosstab data for standard error
drugSTEtumor_bytimepoint=pd.crosstab(index = tumorresponse_stderr_df["Timepoint"], columns = tumorresponse_stderr_df["Drug"], values = tumorresponse_stderr_df["Tumor Volume (mm3) Std Error"],aggfunc='mean')
drugSTEtumor_bytimepoint
```

```python
# Reformat cross-tabulated dataframe to focus on drugs we want to analyze & preview

drugavgtumor_bytimepoint = drugavgtumor_bytimepoint[['Capomulin', 'Infubinol', 'Ketapril', 'Placebo']]
drugavgtumor_bytimepoint.head()
```

```python
# Generate the Plot (with Error Bars)

# Tell matplotlib to create a scatter plot based on drugavgtumor_bytimepoint data frame

formatting = [("Capomulin","o","red"),("Infubinol","^","blue"),("Ketapril","s","green"),("Placebo","d","black")]
legend_labels = ["Capomulin", "Infubinol", "Ketapril", "Placebo"]
for drug, marker, colors in formatting:
    plt.errorbar(drugavgtumor_bytimepoint.index,drugavgtumor_bytimepoint[drug], yerr=drugSTEtumor_bytimepoint[drug],fmt=marker,ls = "--", color=colors)

# Add title,legend, and appropriate axes titles to chart
plt.title("Tumor Response to Treatment")
plt.legend(legend_labels, loc="best")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")

# Save the Figure
plt.savefig(os.path.join("figure_files","Tumor Response to Treatment.png"))

# Show the figure
plt.grid()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint AND convert to DataFrame
metsites_avg_df = combined_df.loc[:,['Drug','Timepoint','Metastatic Sites']].groupby(['Drug','Timepoint']).mean().reset_index().rename(columns = {"Metastatic Sites": "Metastatic Sites Average"})

# Preview DataFrame
metsites_avg_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
# Convert to DataFrame

metsites_stderr_df = combined_df.loc[:,['Drug','Timepoint','Metastatic Sites']].groupby(['Drug','Timepoint']).sem().reset_index().rename(columns = {"Metastatic Sites": "Metastatic Sites Std Error"})

# Preview DataFrame

metsites_stderr_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames -- for average
drugmssites_bytimepoint=pd.crosstab(index = metsites_avg_df["Timepoint"], columns = metsites_avg_df["Drug"], values = metsites_avg_df["Metastatic Sites Average"], aggfunc='mean')

# Preview that Reformatting worked
drugmssites_bytimepoint
```

```python
# Minor Data Munging to Re-Format the Data Frames -- for standard error
drugmssites_SEM_bytimepoint=pd.crosstab(index = metsites_stderr_df["Timepoint"], columns = metsites_stderr_df["Drug"], values = metsites_stderr_df["Metastatic Sites Std Error"], aggfunc='mean')

# Preview that Reformatting worked
drugmssites_SEM_bytimepoint
```

```python
# Generate the Plot (with Error Bars)

for drug, marker, colors in formatting:
    plt.errorbar(drugmssites_bytimepoint.index,drugmssites_bytimepoint[drug], yerr=drugmssites_SEM_bytimepoint[drug],fmt=marker,ls = "--", color=colors)

    
plt.title("Metastatic Spread During Treatment")
plt.legend(legend_labels, loc="best")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Met. Sites")

# Save the Figure
plt.savefig(os.path.join("figure_files","Metastatic Spread During Treatment.png"))

# Show the Figure
plt.grid()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric) AND convert to data frame
mouse_count_df = combined_df.loc[:,['Drug','Timepoint','Mouse ID']].groupby(['Drug','Timepoint']).count().reset_index().rename(columns = {"Mouse ID": "Mouse Count"})

# Preview DataFrame

mouse_count_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mousecount_bytimepoint=pd.crosstab(index = mouse_count_df["Timepoint"], columns = mouse_count_df["Drug"], values = mouse_count_df["Mouse Count"], aggfunc='mean')

# Preview the Data Frame
mousecount_bytimepoint.head()
```

```python
# Generate the Plot (Accounting for percentages)

for drug, marker, colors in formatting:
    plt.plot(mousecount_bytimepoint.index,(mousecount_bytimepoint[drug]/(mousecount_bytimepoint[drug][0]))*100,marker=marker,ls = "--", color=colors)

    
plt.title("Survival During Treatment")
plt.legend(legend_labels,loc="best")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")

# Save the Figure
plt.savefig(os.path.join("figure_files","Survival During Treatment.png"))

# Show the Figure
plt.grid()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
percent_change = ((drugavgtumor_bytimepoint.iloc[-1] / drugavgtumor_bytimepoint.iloc[0])-1)*100

# Display the data to confirm
percent_change
```

```python
# Store all Relevant Percent Changes into a Tuple
drug_list = drugavgtumor_bytimepoint.columns

percentChange_list = [percent_change[drug] for drug in drug_list]
tuple(percentChange_list)
```

```python
# Create bar graph for tuple of percent change of tumor volume

passing_rate = percent_change <= 0

plt.bar(drug_list,percentChange_list,width=-1,align="edge",color=passing_rate.map({True:"g",False:"r"}))

# format plot
plt.ylim(-30,70)
plt.ylabel("% Tumor Volume Change")
plt.title("Tumor Change Over 45 Day Treatment")

# add data labels to each bar
# def autolabel(rects):
#     for rect in rects:
#         height = rect.get_height()
#         ax.text(rect.get_x() + rect.get_width()/2., 
#                 *height,
#                 '%d' % int(height),
#                 ha='center', va='bottom')

for x,y in zip(drug_list,percentChange_list):
    label = "{:.0f}%".format(y)
    plt.annotate(label,(x,y),textcoords="offset points",xytext=(0,10),ha="center")

# Save the Figure
plt.savefig(os.path.join("figure_files","Tumor Change Over 45 Day Treatment.png"))

# Show the Figure
plt.grid()
```
